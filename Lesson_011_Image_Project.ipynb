```python
import numpy as np
import matplotlib.pyplot as plt
image = np.zeros((10, 10))


image [0, :] = 255
mask = np.zeros((10,10), dtype=bool)
mask[4:7, 4:7] = True
image[mask] = 150


print("Numerical Array:\n", image)
plt.imshow(image, cmap= 'gray')
plt.title("Numpy Image Project")
plt.show()
```
