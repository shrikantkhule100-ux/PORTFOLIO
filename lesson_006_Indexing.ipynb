```python
import numpy as np
matrix = np.array([[10,20,30],
                  [40,50,60],
                  [70,80,90]])
middle_value = matrix[1,1]

first_row = matrix[0,:]
last_column = matrix[:,2]
print("full matrix:\n",matrix)
print(" The middle value",middle_value)
print("The first row",first_row)
print("The last column",last_column)

```
