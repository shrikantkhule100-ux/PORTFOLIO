```python
numbers = [1,2,3,4,5]
doubled_numbers = []
for x in numbers:
  doubled_numbers.append(x*2)
print("the orignal_array", numbers)
print("manual result",doubled_numbers)


import numpy as np
numbers_array = np.array ([1,2,3,4,5])
doubled_array = numbers_array *2

print("the orignal_array", numbers_array)
print("the numpy array", doubled_array)
```
