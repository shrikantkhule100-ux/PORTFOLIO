```python
# manual way to find number>50
data=[20,60,30,80,40]
filter=[]
for x in data:
  if x > 50:
    filter.append(x)
print(filter)

# proffetional way
import numpy as np
data = np.array([20,60,30,80,40])
filter=data>50
print(data[filter])


```
